# MongoDB

MongoDB is a popular open-source document-oriented database system. It falls under the category of NoSQL (non-relational) databases, which means it does not rely on the traditional tabular structure of rows and columns used in relational databases. MongoDB stores data in a flexible, JSON-like format called BSON (Binary JSON), which allows for dynamic schemas and easy representation of complex data structures.

```python
# Replace 'products' with the actual name of your collection storing products
products_collection = db.products

# Replace 'users' with the actual name of your collection storing users
users_collection = db.users

# Replace 'categories' with the actual name of your collection storing categories
categories_collection = db.categories

# Fetch products with user name and category name
cursor = products_collection.find()

for product in cursor:
    # Fetch user name using the user ID
    user_id = product.get("user_id")
    user = users_collection.find_one({"_id": user_id})
    user_name = user.get("name") if user else "Unknown User"

    # Fetch category name using the category ID
    category_id = product.get("category_id")
    category = categories_collection.find_one({"_id": category_id})
    category_name = category.get("name") if category else "Unknown Category"

    # Print or process the information as needed
    print(f"Product: {product.get('name')}, User: {user_name}, Category: {category_name}")
```

## Aggreagation

```python
# Moyenne Age
result=products_collection.aggregate(
   [
     {
       "$group":
         {
           "_id": "null",
           "avgAge": { "$avg": "$price" }
         }
     }
   ]
)


# Check it
print(list(result))
```

```python
# Somme Age
result=products_collection.aggregate(
   [
     {
       "$group":
         {
           "_id": "null",
           "somme": { "$sum": "$price" }
         }
     }
   ]
)


# Check it
print(list(result))
```

## Index

```python
# List indexes of the collection
indexes = collection.index_information()
print("List of indexes:", indexes)

# Index the "name" field in the collection "users"
index_name = "_name_index"
collection.create_index([("name", pymongo.ASCENDING)], name=index_name)
print(f"Index '{index_name}' created on the 'name' field.")

# Drop the previously created index
collection.drop_index(index_name)
print(f"Index '{index_name}' dropped.")

# Check the usage of indexes when finding a document with "_id" field equal to "656c78286fd419c668d57a8b"
query = {"_id": "656c78286fd419c668d57a8b"}
result = collection.find(query).explain()
print("Index usage for query:", result)

# Find the document with "_id" field equal to "656c78286fd419c668d57a8b" and explain the query execution
result_with_explain = collection.find(query).explain()
print("Query execution details:", result_with_explain)
```

```python
# Example: Aggregation pipeline to get a virtual "view"
pipeline = [
    {
        "$lookup": {
            "from": "categories",
            "localField": "category_id",
            "foreignField": "_id",
            "as": "category"
        }
    },
    {
        "$lookup": {
            "from": "users",
            "localField": "user_id",
            "foreignField": "_id",
            "as": "user"
        }
    },
    {
        "$unwind": "$category"
    },
    {
        "$unwind": "$user"
    },
    {
        "$project": {
            "product_name": "$name",
            "price": 1,
            "category_name": "$category.name",
            "user_name": "$user.name"
        }
    }
]

# Use the aggregation pipeline
result = list(products_collection.aggregate(pipeline))

# Now 'result' contains documents with product details, category name, and user name
print(result)
```
